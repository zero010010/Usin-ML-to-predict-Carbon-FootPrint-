* The recycling methods and cooking ingredients likely have an influence on the target variable (carbon footprint). Understanding their individual effects could provide insights.

```python

df = pd.read_csv('/Users/miguelopez/Desktop/4-Proyecto ML/src/data_raw/Carbon Emission.csv')
# 'Recycling' and 'Cooking_With' columns from string to lists
df['Recycling'] = df['Recycling'].apply(ast.literal_eval)
df['Cooking_With'] = df['Cooking_With'].apply(ast.literal_eval)

# Flatten 
df['Recycling'] = df['Recycling'].apply(lambda x: ','.join(x))
df['Cooking_With'] = df['Cooking_With'].apply(lambda x: ','.join(x))

# dummy variables
recycling_dummies = df['Recycling'].str.get_dummies(sep=',')
cooking_with_dummies = df['Cooking_With'].str.get_dummies(sep=',')

# Concatenate 
df = pd.concat([df.drop(['Recycling', 'Cooking_With'], axis=1), recycling_dummies, cooking_with_dummies], axis=1)



```

```python
# Target Hist

sns.set_style("darkgrid")
plt.figure(figsize=(10, 10), dpi=50)
sns.histplot(df['CarbonEmission'], kde=True, color='green', bins=30)
plt.title('Distribution of Carbon Emissions')
plt.xlabel('Carbon Emission')
plt.ylabel('Frequency')
plt.show()
```

<span style="color:blue">Right skeked distribution:</span> <br> 
Single peak, long tail to the right, very high emission values pulling the mean (2269) above the median (2080).

<span style="color:blue">Concentration Cluster:</span>
 <br>
Between 1500-3000, suggesting most people have moderate emissions in this range.

<span style="color:blue">Minimum Value:</span>
 <br>306, indicating there are some very low emitters.

<span style="color:blue">Maximum Value:</span>
 <br>8377, showing there are also some extreme high emitters.

<span style="color:blue">Outliers:</span>
 <br> 

The shape is unimodal and fairly smooth, no major outliers.<br>

*However, on the upper end, values above approximately 6000 could potentially be considered outliers. The bulk of the distribution lies between 1500-3000, so emissions higher than 6000 seem unusually high. Specifically, there is a cluster of points between 7000-8000 that protrudes quite far past the rest of the distribution.*



```python
#  boxplots showing relationship between diet, transport, energy efficiency vs carbon emission .

# figure and axes
fig, axes = plt.subplots(nrows=3, figsize=(10, 18))

# Dictionary to map axes and columns 
plots = {axes[0]: ['Diet', 'CarbonEmission'], 
         axes[1]: ['Transport', 'CarbonEmission'],
         axes[2]: ['Energy efficiency', 'CarbonEmission']}

# boxplots
for ax, (x, y) in plots.items():
    sns.boxplot(ax=ax, x=x, y=y, data=df, palette="coolwarm")
    ax.set_xlabel(x)
    ax.set_ylabel('Carbon Emission')

# Set titles    
axes[0].set_title('By Diet')
axes[1].set_title('By Transport') 
axes[2].set_title('By Energy Efficiency')

# Adjust layout and show plot
plt.tight_layout()
plt.show()
```

```python
# 
# Diet type vs Transport / Carbon Emissions
plt.figure(figsize=(14, 8), dpi=200)
sns.barplot(x='Diet', y='CarbonEmission', hue='Transport', data=df, palette='Blues')
plt.title('Carbon Emission vs Diet type and Transport Method')
plt.xlabel('Diet')
plt.ylabel('Carbon Emission')
plt.legend(title='Transport Method', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
